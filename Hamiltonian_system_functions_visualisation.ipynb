```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
omega = 1
lam = 1

#Grid
x = np.linspace(-3, 3, 300)
y = np.linspace(-3, 3, 300)

X, Y = np.meshgrid(x, y)

#Potential energy
V = (omega / 2) * X**2 - (lam / 2) * Y**2

#Figure
fig = plt.figure(figsize = (8,6))
ax = fig.add_subplot(111, projection = '3d')#1*1 grid, 1st subplot

#Surface
surf = ax.plot_surface(
    X, Y, V,
    cmap = 'viridis',
    edgecolor = 'none'
)

#Labels
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$y$')
ax.set_zlabel(r'$V(x,y)$') #This is hidden but seems not a big deal.

ax.set_title(r'$V(x,y)$')

#Colorbar
fig.colorbar(surf, shrink = 0.6)#The potential energy V(x, y) is in colour scale.

#Viewing angle
ax.view_init(elev = 30, azim = -120)#elevation vs azimuth angles

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
omega = 1
lam = 1

#Grid
x = np.linspace(-3, 3, 400)
y = np.linspace(-3, 3, 400)

X, Y = np.meshgrid(x, y)

#Potential energy
V = (omega / 2) * X**2 - (lam / 2) * Y**2

#Figure
fig, ax = plt.subplots(figsize = (9,7))

#Colour background
cf = ax.contourf(
    X, Y, V,
    levels = 50,
    cmap = 'viridis'
)

#Equipotential lines
ax.contour(
    X, Y, V,
    levels = 10,
    colors = 'black',
    linewidths = 1
)

#Diagonal equipotential lines
ax.plot([-3, 3], [3, -3], color='black')
ax.plot([-3, -3], [3, 3], color='black')

#Example Γ
ax.plot(
    np.linspace(-2, 1.5, 200),
    np.zeros(200),
    color = 'red',
    linewidth = 3,
    label = r'$\Gamma$'
)

#Example γ
traj_x = [0.1, 0.6, 1.2, 1.5, 0.3, -0.5, -1.0, -1.3, -1.4, -1.3, -1.2, -1.1]
traj_y = [0.0, 0.005, 0.01, 0.02, 0.15, 0.3, 0.7, 1.2, 1.8, 2.3, 2.8, 3.4]

ax.plot(
    traj_x,
    traj_y,
    color = 'orange',
    linewidth = 3,
    label = r'$\gamma$'
)

#Labels
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$y$')
ax.set_title(r'$V(x,y)$')

#Limits
ax.set_xlim(-3, 3)
ax.set_ylim(-3, 3)

#Colorbar
fig.colorbar(cf, ax=ax)

#Legend
ax.legend()

plt.show()
```

```python
#3.Hamiltonian Monte Carlo (HMC). 
#(According to https://www.cs.helsinki.fi/u/ahonkela/teaching/compstats1/book/hamiltonian-monte-carlo-hmc.html#hmc-theory)
#We illustrate HMC on a spherical density
```

```python
from IPython.display import display, Math

display(Math(
    r'\log \pi(\theta) = -b(\|\theta\|_2 - r)^2,'
))
```

```python
#which reduces to a circle in 2D. b and r are positive, -log pi(theta) denotes the potential energy of the system,
#theta is the position, and r is the momentum variable.
```

```python
#Set b = 5 and r = 10.
import autograd.numpy as np
import autograd.numpy.random as npr
import autograd
import matplotlib.pyplot as plt

np.random.seed(100)

def hmc(theta0, M, target, epsilon, L):
    thetas = np.zeros([M, len(theta0)])
    Nabla = autograd.grad(target)
    theta = np.copy(theta0)
    g = Nabla(theta)  #set gradient using initial theta
    log_pi = target(theta)  #set objective function
    a = 0 #set the Metropolis–Hastings acceptance probability
    for m in range(M): #draw M samples
        r = npr.normal(size=theta.shape)  #initial momentum is Normal(0,1)
        H = r.T @ r / 2 - log_pi   #evaluate H(theta,p), which equals the sum of the kinetic energy and the potential energy
        thetanew = np.copy(theta)
        gnew = np.copy(g)
        for l in range(L): #make L 'leapfrog' steps
            r = r + epsilon * gnew / 2   #make half-step in p
            thetanew = thetanew + epsilon * r    #make step in theta
            gnew = Nabla(thetanew)           #find new gradient
            r = r + epsilon * gnew / 2   #make half-step in p
        log_pinew = target(thetanew)   #find new value of H
        Hnew = r.T @ r / 2 - log_pinew
        log_a = H - Hnew    #decide whether to accept
        if np.log(npr.rand()) < log_a:
            g = gnew
            theta = thetanew
            log_pi = log_pinew
            a += 1
        thetas[m,:] = theta
    print('Acceptance rate:', a/M)
    return thetas

def sphere(theta):
    return -5 * (np.sqrt(np.sum(theta**2))-10)**2

samples = hmc(np.array([3.0, 0.0]), 200, lambda theta: sphere(theta), 0.2, 50)
```

```python
samples = samples[samples.shape[0]//2:]
plt.plot(samples[:,0], samples[:,1])
plt.show()
```
