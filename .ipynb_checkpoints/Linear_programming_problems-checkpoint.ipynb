```python
#1 A company wants to purchase at most 1800 units of a product. There are two types of the product available: 
## product 1 and product 2.Product 1 occupies 2 ft3, costs £12 and the company makes a profit of £3. Product 2 
## occupies 3 ft3, costs £15 and the company makes a profit of £4. The company wants to maximise its profit. The 
## budget is £15 000 and the warehouse has 3 000 ft3 for the product.
```

```python
# Change the kernel to SageMath, download it if you haven't.
```

```python
p = MixedIntegerLinearProgram(maximization = True,solver = "GLPK")
```

```python
# x[i] (i = 1,2) denote the quantity of product i
x = p.new_variable(real = True, nonnegative = True)
```

```python
p.set_objective(3*x[1] + 4*x[2]) #profit
p.add_constraint(x[1] + x[2] <= 1800) #number of units
p.add_constraint(12*x[1] + 15*x[2] <= 15000) #budget in pounds
p.add_constraint(2*x[1] + 3*x[2] <= 3000) #volume in ft3
p.add_constraint(x[1] >= 0) #nonnegativity of the quantity
p.add_constraint(x[2] >= 0) #nonnegativity of the quantity
```

```python
p.show()
```

```python
p.solve() #maximum profit in pounds
```

```python
for i, v in p.get_values(x).items():
    print('x_%s = %s' % (i,v))
```

```python
# The optimal profit is £4000, with 0 units of product 1 and 1000 units of product 2.
```
